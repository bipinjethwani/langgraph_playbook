```python
from pydantic import BaseModel

class User(BaseModel):
    name: str
    age: int
    email: str

# This works ✅
user = User(name="Bipin", age=25, email="bipin@example.com")
user
```

```python
# This FAILS with a clear error ❌
user = User(name="Bipin", age="twenty-five", email=12345)
# ValidationError: age must be an integer
```

```python
from pydantic import BaseModel

class Fruit(BaseModel):
    name: str
    price: float

# Create an instance
apple = Fruit(name="apple", price=1.5)

# Access like normal attributes
print(apple.name)   # "apple"
print(apple.price)  # 1.5
```

```python
# You pass a string "1.5", Pydantic converts to float
apple = Fruit(name="apple", price="1.5")
print(apple.name)       # 1.5 (float, not string!)
print(type(apple.price)) # <class 'float'>
print(apple.price) # <class 'float'>
```
